```python
#######################################################################
#
# An example of a Combined chart in XlsxWriter.
#
# SPDX-License-Identifier: BSD-2-Clause
# Copyright 2013-2022, John McNamara, jmcnamara@cpan.org
#
from xlsxwriter.workbook import Workbook

workbook  = Workbook('MF_target.xlsx')
worksheet={}
j=1
for i in dict123.keys(): 
    worksheet[i] = workbook.add_worksheet(i[:30])

    # Add a format for the headings.
    bold = workbook.add_format({'bold': True})

    # Add the worksheet[i] data that the charts will refer to.
    headings = [i, 'No of policies', 'MF TARGET']

    data=dict123[i][['bin','count','percentage']].T.values

    worksheet[i].write_row('A1', headings, bold)
    worksheet[i].write_column('A2', data[0])
    worksheet[i].write_column('B2', data[1])
    worksheet[i].write_column('C2', data[2])


    # Create a new column chart. This will use this as the primary chart.
    column_chart2 = workbook.add_chart({'type': 'column'})

    # Configure the data series for the primary chart.
    column_chart2.add_series({
        'name':       '='+i[:30]+'!$B$1',
        'categories': '='+i[:30]+'!$A$2:$A$'+str(dict123[i].shape[0]+1),
        'values':     '='+i[:30]+'!$B$2:$B$'+str(dict123[i].shape[0]+1),
        'data_labels': {'value': True},
        'major_gridlines' :{'visible': False}
    })

    # Create a new column chart. This will use this as the secondary chart.
    line_chart2 = workbook.add_chart({'type': 'line'})

    # Configure the data series for the secondary chart. We also set a
    # secondary Y axis via (y2_axis). This is the only difference between
    # this and the first example, apart from the axis label below.
    line_chart2.add_series({
        'name':       '='+i[:30]+'!$C$1',
        'categories': '='+i[:30]+'!$A$2:$A$'+str(dict123[i].shape[0]+1),
        'values':     '='+i[:30]+'!$C$2:$C$'+str(dict123[i].shape[0]+1),
        'y2_axis':    True,
        'data_labels': {'value': True}
    })

    # Combine the charts.
    column_chart2.combine(line_chart2)

    # Add a chart title and some axis labels.
    column_chart2.set_title({  'name': i})
    column_chart2.set_x_axis({ 'name': 'bins'})
    column_chart2.set_y_axis({ 'name': 'Owner counts','major_gridlines': {
        'visible': False},'minor_gridlines': {
        'visible': False}})

    # Note: the y2 properties are on the secondary chart.
    line_chart2.set_y2_axis({'name': 'Event Rate'})
    
    # Position the legend at the top of the chart.
    column_chart2.set_legend({'position': 'bottom'})

    # Insert the chart into the worksheet[i]
    worksheet[i].insert_chart('F2', column_chart2)
    j=j+1

workbook.close()
```
