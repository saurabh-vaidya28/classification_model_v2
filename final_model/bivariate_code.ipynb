```python
import pandas as pd
import numpy as np

import warnings
warnings.filterwarnings("ignore")

pd.set_option("display.max_columns", 600)
pd.set_option("display.max_rows", 600)
```

```python
df_train = pd.read_csv(r'Train_v1.2.csv',sep=',',encoding='utf-8',parse_dates=True)
df_train.head()
```

```python
for i in df_train.columns:
    if df_train[i].dtypes == 'object':
        df_train[i] = df_train[i].str.upper()
        df_train[i] = df_train[i].str.strip()
```

```python
# converting float to integer if column have 2 unique value
for i in df_train.columns:
    if df_train[i].dtypes == 'float':
        if df_train[i].nunique() <= 2:
            df_train[i] = df_train[i].astype(int)
```

```python
# Check the event rate
df_train_er = pd.DataFrame(df_train['Default'].value_counts())
df_train_er['Event Rate'] = (df_train_er['Default'] / df_train_er['Default'].sum())*100
df_train_er
```

```python
# !pip install --upgrade matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
import scorecardpy as sc
dict1={}
%matplotlib inline

error_cols=[]
df123=pd.DataFrame()
dict123={}

for col in df_train.columns:
    try:

        Max = df_train[[col, 'Default']]
        bins = sc.woebin(Max, y = 'Default')
        bins[col]['percentage']=round(bins[col]['badprob'],3)
        a = bins[col]
        a.drop(['count_distr', 'badprob',
       'woe', 'bin_iv', 'total_iv', 'breaks', 'is_special_values'], axis = 1, inplace = True)
        na_df = pd.DataFrame(columns=['variable', 'bin', 'good', 'bad', 'count', 'percentage'])
        a = na_df.append(a)
        a.reset_index(inplace=True, drop=True)
#         print(a)
        df123=pd.concat([df123,a],axis=0)
        dict123[col]= a
    except:
        print('error',col)
        error_cols.append(col)
```

```python
df123.to_excel('numerical_bi_variate.xlsx', index = False)
```

```python
from xlsxwriter.workbook import Workbook

workbook  = Workbook('Default.xlsx')
worksheet={}
j=1
for i in dict123.keys(): 
    try:
        
        worksheet[i] = workbook.add_worksheet(i[:30])

        # Add a format for the headings.
        bold = workbook.add_format({'bold': True})

        # Add the worksheet[i] data that the charts will refer to.
        headings = [i, 'No of policies', 'TARGET', 'GOOD', 'BAD']

        data=dict123[i][['bin','count','percentage', 'good', 'bad']].T.values

        worksheet[i].write_row('A1', headings, bold)
        worksheet[i].write_column('A2', data[0])
        worksheet[i].write_column('B2', data[1])
        worksheet[i].write_column('C2', data[2])
        worksheet[i].write_column('D2', data[3])
        worksheet[i].write_column('E2', data[4])


        # Create a new column chart. This will use this as the primary chart.
        column_chart2 = workbook.add_chart({'type': 'column'})

        # Configure the data series for the primary chart.
        column_chart2.add_series({
            'name':       '='+i[:30]+'!$B$1',
            'categories': '='+i[:30]+'!$A$2:$A$'+str(dict123[i].shape[0]+1),
            'values':     '='+i[:30]+'!$B$2:$B$'+str(dict123[i].shape[0]+1),
            'data_labels': {'value': True},
            'major_gridlines' :{'visible': False}
        })

        # Create a new column chart. This will use this as the secondary chart.
        line_chart2 = workbook.add_chart({'type': 'line'})

        # Configure the data series for the secondary chart. We also set a
        # secondary Y axis via (y2_axis). This is the only difference between
        # this and the first example, apart from the axis label below.
        line_chart2.add_series({
            'name':       '='+i[:30]+'!$C$1',
            'categories': '='+i[:30]+'!$A$2:$A$'+str(dict123[i].shape[0]+1),
            'values':     '='+i[:30]+'!$C$2:$C$'+str(dict123[i].shape[0]+1),
            'y2_axis':    True,
            'data_labels': {'value': True}
        })

        # Combine the charts.
        column_chart2.combine(line_chart2)

        # Add a chart title and some axis labels.
        column_chart2.set_title({  'name': i})
        column_chart2.set_x_axis({ 'name': 'bins'})
        column_chart2.set_y_axis({ 'name': 'Owner counts','major_gridlines': {
            'visible': False},'minor_gridlines': {
            'visible': False}})

        # Note: the y2 properties are on the secondary chart.
        line_chart2.set_y2_axis({'name': 'Event Rate'})

        # Position the legend at the top of the chart.
        column_chart2.set_legend({'position': 'bottom'})

        # Insert the chart into the worksheet[i]
        worksheet[i].insert_chart('F2', column_chart2)
        j=j+1
    except:
        pass
workbook.close()
```
